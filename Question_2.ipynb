## Load NYC filtered dataset

```python
import pandas as pd
import numpy as np
```

```python
# Load it into a Dataframe using pandas
path = 'nyc311_filter.csv'
df = pd.read_csv(path)
df.head()
```

# What are the anomalous complaints per department for each zip code with respect to the total?

```python
# Group complaints
call_freq = df['complaint_type'].groupby(df['complaint_type']).count(
).sort_values(ascending=False).reset_index(name='total_count')
```

```python
call_freq.head(10)
```

```python
# Group complaints as per department and zipcode
df_count_zip = df.groupby(['complaint_type','incident_zip', 'agency']).size(
).sort_values(ascending=False).reset_index(name="complaint_count")
df_count_zip.head(10)
```

```python
mergedStuff = pd.merge(call_freq, df_count_zip, on=['complaint_type'], how='inner')
mergedStuff['Percentage'] = (mergedStuff.complaint_count / mergedStuff.total_count) * 100
mergedStuff.head(10)
```

```python
mergedStuff.to_csv('percentage.csv', index=False)
```
